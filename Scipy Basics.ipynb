```python
cov=np.array(([1,0.8],[0.8,3]))

#Variance 1 in first dimension
#Variance 3 in second dimension
# covariance between them is 0.8
```

```python
from scipy.stats import multivariate_normal as mvn
```

```python
mean=np.array([0,2])
```

```python
r=mvn.rvs(mean=mean,cov=cov,size=1000)
a=r[:,0]
b=r[:,1]
plt.scatter(a,b)
plt.axis('equal')
```

```python
r=np.random.multivariate_normal(mean=mean,cov=cov,size=100000)
a=r[:,0]
b=r[:,1]
plt.scatter(a,b)
plt.axis('equal')
```
